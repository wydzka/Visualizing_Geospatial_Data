No.1
- Tampilkan rumah paling mahal
- Tandai dengan icon pin berwarna merah
- Beri tooltip berupa "Rumah Termahal"

```python
df= pd.read_csv('melb_data.csv')
df.head()
```

```python
df['Price'].max()
```

```python
df_max = df[(df['Price'] == df['Price'].max()) & (df['Type'] == 'h')]
df_max
```

```python
longtitud = df_max['Longtitude'].values[0]
longtitud
```

```python
lattitude = df_max['Lattitude'].values[0]
lattitude
```

```python
## buat peta peta new york
# buat peta New York
melb_map = folium.Map(location=[lattitude, longtitud], zoom_start=12)
melb_map

# buat merker untuk liberty statu
# buat marker untuk Liberty Statue
melb_marker = folium.Marker(
    location=[lattitude, longtitud],
    tooltip='ini rumah termahal',
    popup='35 Bevis St',
    icon= folium.Icon(
                color='red',
                icon='home')
)


# # buat marker untuk wilayah broklin
# brooklyn_area = folium.Circle(
#     location=[longtitud, lattitude],
#     tooltip=' Area rawan Banjir',
#     popup='brooklyn',
#     radius= 2500



## memasukan marker ke map
melb_marker.add_to(melb_map)
# brooklyn_area.add_to(melb_map)

# menampilkan map
melb_map
```
